### **Overview**
This notebook performs manual mathematical operations as taught in **Day 45: Math-For-ML** to verify the data's statistical properties.

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
df = pd.read_csv(url).dropna(subset=['age', 'fare'])

# 1. Manual Mean and Std
ages = df['age'].values
manual_mean = np.sum(ages) / len(ages)
manual_std = np.sqrt(np.sum((ages - manual_mean)**2) / len(ages))

print(f"Manual Mean Age: {manual_mean:.2f}")
print(f"Manual Std Dev: {manual_std:.2f}")
```

```python
# 2. Cosine Similarity
# Comparing the first two passengers based on Age and Fare
p1 = df[['age', 'fare']].iloc[0:1]
p2 = df[['age', 'fare']].iloc[1:2]
sim = cosine_similarity(p1, p2)

print(f"Cosine Similarity between Passenger 1 and 2: {sim[0][0]:.4f}")
```

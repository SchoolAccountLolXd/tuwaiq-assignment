```python
# Libraries re-imported for Math tasks
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 1. Load Data
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
df = pd.read_csv(url).dropna(subset=['age', 'fare'])

# 2. Manual Math (Ref: Course Day 45)
# AI Citation: ChatGPT helped format the manual standard deviation formula
ages = df['age'].values
m_mean = np.sum(ages) / len(ages)
m_std = np.sqrt(np.sum((ages - m_mean)**2) / len(ages))

print(f"Manual Mean Age: {m_mean:.2f}")
print(f"Manual Std Dev: {m_std:.2f}")

# 3. Cosine Similarity (Ref: GeeksforGeeks)
p1 = df[['age', 'fare']].iloc[0:1]
p2 = df[['age', 'fare']].iloc[1:2]
sim = cosine_similarity(p1, p2)
print(f"Cosine Similarity (Passenger 1 vs 2): {sim[0][0]:.4f}")
```
